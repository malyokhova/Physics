```python
e = 1.6 * math.pow(10, -19) # эВ в Дж
h = 6.63 * math.pow(10, -34)
c = 3 * math.pow(10, 8)
dj = 6.2 * math.pow(10, 18) # Дж в эВ

wl_1 = 520 * math.pow(10, -9) # м
wl_2 = 565 * math.pow(10, -9) # м
wl_3 = 590 * math.pow(10, -9) # м

U_b_1 = 8.0
U_b_2 = 7.0
U_b_3 = 6.0
U_a = 0.5

E_1 = e * (U_b_1 - U_a) # Дж
E_2 = e * (U_b_2 - U_a)
E_3 = e * (U_b_3 - U_a)

print('энергии 10^19', E_1 * math.pow(10, 19), E_2 * math.pow(10, 19), E_3 * math.pow(10, 19))

v_1 = c / wl_1 - e * U_a / h
v_2 = c / wl_2 - e * U_a / h
v_3 = c / wl_3 - e * U_a / h

print('пороги фотоэффекта 10^14', v_1 * math.pow(10, -14), v_2 * math.pow(10, -14), v_3 * math.pow(10, -14))

# print('длины волн порогов фотоэффекта, нм', c / v_1 * math.pow(10, 9), c / v_2 * math.pow(10, 9), c / v_3 * math.pow(10, 9))
```

```python
c = 3 * math.pow(10, 8)
e = 1.6 * math.pow(10, -19)
h = 6.63 * math.pow(10, -34)
wl = [430, 470, 520, 565, 590, 660, 700, 860]
I = [3.94, 3.77, 3.22, 2.56, 2.5, 1.55, 0.7, 0.69]
v = [c / w * math.pow(10, 9) for w in wl]

plt.minorticks_on()
plt.grid(True, which='both')
plt.grid(which='minor', color='gray', linestyle=':')

plt.plot(v, I, color='blue')

v_approx = [v[6], v[5]]
I_approx = [I[6], I[5]]

a, b = np.polyfit(v_approx, I_approx, 1)
v_0 = -b / a
wl_0 = c / v_0 * math.pow(10, 9)
print('Порог фотоэффекта 10^14', v_0 * math.pow(10, -14))
print('Работа выхода', h * v_0 * dj)
print('Материал оксид бария')

v = [-b / a, v[6], v[5]]

plt.plot(v, [a * v_ + b for v_ in v], color='red')

plt.title("Спектральная характеристика")
plt.xlabel(r'$\vartheta$, Гц')
plt.ylabel("I, A")

plt.show()
```
